```python
import numpy as np
import matplotlib.pyplot as plt
from general_functions import *
```

```python
def f(r, alpha=0.8): return r**(-alpha)

fig, ax = plt.subplots()

ranks = np.linspace(1,11,50)
for alpha in [0.1, 0.5, 1., 2.]:
    ax.plot(ranks, f(ranks, alpha=alpha), label=f'{alpha:.1f}', alpha=0.5)

ax.legend()
```

```python
meaning_order = [0, 1, 2, 3, 4]
alpha = 2.
M = 5

meaning_ranks = np.array(meaning_order) + 1
meaning_freqs = meaning_ranks ** (-alpha)
meaning_freqs /= np.sum(meaning_freqs)

intends = []

print(meaning_freqs)

for i in range(1000):
    intend = np.random.choice(M, p=meaning_freqs)
    intends.append(intend)

print(intends)
plt.hist(intends, bins=np.arange(6), width=0.8)
plt.show()
```

```python
S = 5
alpha = 0.8
prob_array = np.arange(1, S+1, dtype=float)
prob_array /= np.sum(prob_array)
print(prob_array)

signal_ranks = np.arange(1, S+1) # unchanged order of signals, ranked by ease of use
signal_freqs = signal_ranks ** (-alpha)
print(signal_freqs)
total_pdf = signal_freqs * prob_array
total_pdf /= np.sum(total_pdf)
print(total_pdf)

signals = []

for i in range(10000):
    signal = np.random.choice(S, p=total_pdf)
    signals.append(signal)

plt.hist(signals, bins=np.arange(S+1), width=0.8, density=True)
plt.show()
```
